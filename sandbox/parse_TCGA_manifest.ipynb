## Post Request to Download Multiple Files 
This example uses a Python list to specify a set of file UUIDs. The list in the example was populated manually but could potentially be populated programmatically from an external list or API call.

```python
import requests
import json
import re

data_endpt = "https://api.gdc.cancer.gov/data"

#list of uuids can be manually entered here or programatically though the API
ids = ['c3df46a9-85d1-45d4-954a-825313d4a26d']

params = {"ids": ids}

response = requests.post(data_endpt,
                        data = json.dumps(params),
                        headers={
                            "Content-Type": "application/json"
                            })

response_head_cd = response.headers["Content-Disposition"]

file_name = re.findall("filename=(.+)", response_head_cd)[0]

with open(file_name, "wb") as output_file:
    output_file.write(response.content)
```

## Downloading a Set of Files Based on a Filter 

Here a list of files based on a set of filters are downloaded. File UUIDs are retrieved based on the filters. These UUIDs are then passed to the data endpoint to download the correct files.

Link to fields: https://docs.gdc.cancer.gov/API/Users_Guide/Appendix_A_Available_Fields/

```python
##works for "cases.submitter_id" in the files_endpoint "https://api.gdc.cancer.gov/files"
##can only get 2 "files.data_type" at a time: "Gene Expression Quantification" and "Slide Image"
##cant get "Annotated Somatic Mutation" as well
import requests
import json
import re

files_endpt = "https://api.gdc.cancer.gov/files"

##"field": "value"
#"data_type": "Annotated Somatic Mutation"
#"data_type": "Gene Expression Quantification"
#"data_type": "Slide Image"

filters={
    "op":"or",
    "content":[
        {
        "op":"and",
        "content":[
            {
                "op":"in",
                "content":{
                    "field":"cases.submitter_id",
                    "value":['TCGA-FA-A6HN', 'TCGA-GS-A9TX', 'TCGA-FF-8047']
                }
            },
            {
                "op":"in",
                "content":{
                    "field":"files.data_type",
                    "value":"Slide Image"
                }
            }
        ]
        },
        {
        "op":"and",
        "content":[
            {
                "op":"in",
                "content":{
                    "field":"cases.submitter_id",
                    "value":['TCGA-FA-A6HN', 'TCGA-GS-A9TX', 'TCGA-FF-8047']
                }
            },
            {
                "op":"in",
                "content":{
                    "field":"files.data_type",
                    "value":"Gene Expression Quantification"
                }
            }
        ]
        },
#         {
#         "op":"and",
#         "content":[
#             {
#                 "op":"in",
#                 "content":{
#                     "field":"cases.submitter_id",
#                     "value":['TCGA-FA-A6HN', 'TCGA-GS-A9TX', 'TCGA-FF-8047']
#                 }
#             },
#             {
#                 "op":"in",
#                 "content":{
#                     "field":"files.data_type",
#                     "value":"Annotated Somatic Mutation"
#                 }
#             },
#             {
#                 "op":"in",
#                 "content":{
#                     "field":"files.access",
#                     "value":"open" #check to see if any annotation files are actually open access
#                 }
#             }
#         ]
#         }
    ]
}

# Here a GET is used, so the filter parameters should be passed as a JSON string.

params = {
    "filters": json.dumps(filters),
    "fields": "file_id",
    "format": "JSON",
    "size": "1000"
    }

response = requests.get(files_endpt, params = params)

file_uuid_list = []

# This step populates the download list with the file_ids from the previous query
for file_entry in json.loads(response.content.decode("utf-8"))["data"]["hits"]:
    file_uuid_list.append(file_entry["file_id"])

data_endpt = "https://api.gdc.cancer.gov/data"

params = {"ids": file_uuid_list}

response = requests.post(data_endpt, data = json.dumps(params), headers = {"Content-Type": "application/json"})

response_head_cd = response.headers["Content-Disposition"]

file_name = re.findall("filename=(.+)", response_head_cd)[0]

with open(file_name, "wb") as output_file:
    output_file.write(response.content)
```

```python
import requests
import json
import re

files_endpt = "https://api.gdc.cancer.gov/files"        

filters = {
    "op": "and",
    "content":[
        {
        "op": "in",
        "content":{
            "field":"cases.submitter_id",
            "value":['TCGA-FA-A6HN']#, 'TCGA-GS-A9TX', 'TCGA-FF-8047']
            }
        },
        {
        "op": "and",
        "content":[
            {
            "op": "in",
            "content":{
                "field": "files.data_type",
                "value": ["Annotated Somatic Mutation"]
                }
            },
            {
            "op": "in",
            "content":{
                "field": "files.downstream_analyses.output_files.access",
                "value": ["open"]
                }
            }
        ]
        }
    ]
}

# Here a GET is used, so the filter parameters should be passed as a JSON string.
params = {
    "filters": json.dumps(filters),
    "fields": "file_id",
    "format": "JSON",
    "size": "1000"
    }

response = requests.get(files_endpt, params = params)

an_file_uuid_list = []

# This step populates the download list with the file_ids from the previous query
for file_entry in json.loads(response.content.decode("utf-8"))["data"]["hits"]:
    an_file_uuid_list.append(file_entry["file_id"])

data_endpt = "https://api.gdc.cancer.gov/data"

params = {"ids": an_file_uuid_list}

response = requests.post(data_endpt, data = json.dumps(params), headers = {"Content-Type": "application/json"})

response_head_cd = response.headers["Content-Disposition"]

file_name = re.findall("filename=(.+)", response_head_cd)[0]

with open(file_name, "wb") as output_file:
    output_file.write(response.content)
```
